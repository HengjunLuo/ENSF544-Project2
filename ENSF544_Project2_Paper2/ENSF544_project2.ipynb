```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

covid_data = pd.read_csv("./covid_19_data.csv")
```

# **Preparation of the dataset**

```python
import csv

data = []
names = ["Date", "new Deaths", "new Recovery", "new Confirmed", "total Deaths", "total Recovery", "total Confirmed"]

total_deaths = 0
total_recovery = 0
total_confirmed = 0
start = 0
for index, row in covid_data.iterrows():
    if(row['ObservationDate'] == "03/01/2020"):
        start = 1
    if(row['ObservationDate'] == "05/01/2020"):
        start = 0
    if (row['Country/Region'] == "India" and row['ObservationDate'] == "03/01/2020") or (row['Country/Region'] == "India" and start == 1):
        l = []
        l.append(row['ObservationDate'])
        l.append(row['Deaths'])
        l.append(row['Recovered'])
        l.append(row['Confirmed'])
        total_deaths = total_deaths + row['Deaths']
        total_recovery = total_recovery + row['Recovered']
        total_confirmed = total_confirmed + row['Confirmed']
        l.append(total_deaths)
        l.append(total_recovery)
        l.append(total_confirmed)
        data.append(l)

with open('./covid_19_data_India.csv', 'w', newline = '') as file:
    writer = csv.writer(file)
    writer.writerow(names)
    writer.writerows(data)
```

# **Data Preprocessing**

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler

covid_India_data = pd.read_csv('./covid_19_data_India.csv')

X = np.arange(1,62).reshape(-1,1)
y_deaths = np.array(covid_India_data['new Deaths']).reshape(-1,1)
y_recovery = np.array(covid_India_data['new Recovery']).reshape(-1,1)
y_confirmed = np.array(covid_India_data['new Confirmed']).reshape(-1,1)
y_total_deaths = np.array(covid_India_data['total Deaths']).reshape(-1,1)
y_total_recovery = np.array(covid_India_data['total Recovery']).reshape(-1,1)
y_total_confirmed = np.array(covid_India_data['total Confirmed']).reshape(-1,1)

# new deaths
X_train_new_deaths, X_test_new_deaths, y_train_new_deaths, y_test_new_deaths = train_test_split(X, y_deaths, test_size= 0.4, train_size = 0.6)

#new recovery
X_train_new_recovery, X_test_new_recovery, y_train_new_recovery, y_test_new_recovery = train_test_split(X, y_recovery, test_size= 0.4, train_size = 0.6)

#new confirmed
X_train_new_confirmed, X_test_new_confirmed, y_train_new_confirmed, y_test_new_confirmed = train_test_split(X, y_confirmed, test_size= 0.4, train_size = 0.6)

#total deaths
X_train_total_deaths, X_test_total_deaths, y_train_total_deaths, y_test_total_deaths = train_test_split(X, y_total_deaths, test_size= 0.4, train_size = 0.6)

#total recovery
X_train_total_recovery, X_test_total_recovery, y_train_total_recovery, y_test_total_recovery = train_test_split(X, y_total_recovery, test_size= 0.4, train_size = 0.6)

#total confirmed
X_train_total_confirmed, X_test_total_confirmed, y_train_total_confirmed, y_test_total_confirmed = train_test_split(X, y_total_confirmed, test_size= 0.4, train_size = 0.6)

scalerX1 = StandardScaler()
scalerY1 = StandardScaler()
scaled_X_train_new_deaths = scalerX1.fit_transform(X_train_new_deaths)
scaled_X_test_new_deaths = scalerX1.transform(X_test_new_deaths)
scaled_y_train_new_deaths = scalerY1.fit_transform(y_train_new_deaths)
scaled_y_test_new_deaths = scalerY1.transform(y_test_new_deaths)

scalerX2 = StandardScaler()
scalerY2 = StandardScaler()
scaled_X_train_new_recovery = scalerX2.fit_transform(X_train_new_recovery)
scaled_X_test_new_recovery = scalerX2.transform(X_test_new_recovery)
scaled_y_train_new_recovery = scalerY2.fit_transform(y_train_new_recovery)
scaled_y_test_new_recovery = scalerY2.transform(y_test_new_recovery)

scalerX3 = StandardScaler()
scalerY3 = StandardScaler()
scaled_X_train_new_confirmed = scalerX3.fit_transform(X_train_new_confirmed)
scaled_X_test_new_confirmed = scalerX3.transform(X_test_new_confirmed)
scaled_y_train_new_confirmed = scalerY3.fit_transform(y_train_new_confirmed)
scaled_y_test_new_confirmed = scalerY3.transform(y_test_new_confirmed)

scalerX4 = StandardScaler()
scalerY4 = StandardScaler()
scaled_X_train_total_deaths = scalerX4.fit_transform(X_train_total_deaths)
scaled_X_test_total_deaths = scalerX4.transform(X_test_total_deaths)
scaled_y_train_total_deaths = scalerY4.fit_transform(y_train_total_deaths)
scaled_y_test_total_deaths = scalerY4.transform(y_test_total_deaths)

scalerX5 = StandardScaler()
scalerY5 = StandardScaler()
scaled_X_train_total_recovery = scalerX5.fit_transform(X_train_total_recovery)
scaled_X_test_total_recovery = scalerX5.transform(X_test_total_recovery)
scaled_y_train_total_recovery = scalerY5.fit_transform(y_train_total_recovery)
scaled_y_test_total_recovery = scalerY5.transform(y_test_total_recovery)

scalerX6 = StandardScaler()
scalerY6 = StandardScaler()
scaled_X_train_total_confirmed = scalerX6.fit_transform(X_train_total_confirmed)
scaled_X_test_total_confirmed = scalerX6.transform(X_test_total_confirmed)
scaled_y_train_total_confirmed = scalerY6.fit_transform(y_train_total_confirmed)
scaled_y_test_total_confirmed = scalerY6.transform(y_test_total_confirmed)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt 

performance = []
svr1 = SVR(C=1.0, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma='auto', kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)

l = []
svr1.fit(scaled_X_train_new_deaths, scaled_y_train_new_deaths)
new_death_score = svr1.score(scaled_X_test_new_deaths, scaled_y_test_new_deaths)
y_pred_new_death = svr1.predict(scaled_X_test_new_deaths)
new_death_MSE = mean_squared_error(scaled_y_test_new_deaths, y_pred_new_death)
new_death_RMSE = mean_squared_error(scaled_y_test_new_deaths, y_pred_new_death, squared = False)
l.append(new_death_MSE)
l.append(new_death_RMSE)
l.append(new_death_score)
performance.append(l)

l = []
svr2 = SVR(C=1.0, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma='auto', kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
svr2.fit(scaled_X_train_new_recovery, scaled_y_train_new_recovery)
new_recovery_score = svr2.score(scaled_X_test_new_recovery, scaled_y_test_new_recovery)
y_pred_new_recovery = svr2.predict(scaled_X_test_new_recovery)
new_recovery_MSE = mean_squared_error(scaled_y_test_new_recovery, y_pred_new_recovery)
new_recovery_RMSE = mean_squared_error(scaled_y_test_new_recovery, y_pred_new_recovery, squared = False)
l.append(new_recovery_MSE)
l.append(new_recovery_RMSE)
l.append(new_recovery_score)
performance.append(l)

l = []
svr3 = SVR(C=1.0, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma='auto', kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
svr3.fit(scaled_X_train_new_confirmed, scaled_y_train_new_confirmed)
new_confirmed_score = svr3.score(scaled_X_test_new_confirmed, scaled_y_test_new_confirmed)
y_pred_new_confirmed = svr3.predict(scaled_X_test_new_confirmed)
new_confirmed_MSE = mean_squared_error(scaled_y_test_new_confirmed, y_pred_new_confirmed)
new_confirmed_RMSE = mean_squared_error(scaled_y_test_new_confirmed, y_pred_new_confirmed, squared = False)
pred_X = np.arange(62,122).reshape(-1,1)
pred_X = scalerX3.transform(pred_X)
y_pred3 = svr3.predict(pred_X)
y_pred3 = scalerY3.inverse_transform(y_pred3.reshape(-1,1))
#print(y_pred)

total_y = y_confirmed 
total_y = np.append(total_y, y_pred3)
total_days = np.arange(1,122)
plt.plot(total_days, total_y)
plt.xlabel('Days')
plt.ylabel('Daily Confirmed')
plt.legend()
plt.show()
l.append(new_confirmed_MSE)
l.append(new_confirmed_RMSE)
l.append(new_confirmed_score)
performance.append(l)

l = []
svr4 = SVR(C=1.0, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma='auto', kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
svr4.fit(scaled_X_train_total_deaths, scaled_y_train_total_deaths)
total_death_score = svr4.score(scaled_X_test_total_deaths, scaled_y_test_total_deaths)
y_pred_total_death = svr4.predict(scaled_X_test_total_deaths)
total_death_MSE = mean_squared_error(scaled_y_test_total_deaths, y_pred_total_death)
total_death_RMSE = mean_squared_error(scaled_y_test_total_deaths, y_pred_total_death, squared = False)
l.append(total_death_MSE)
l.append(total_death_RMSE)
l.append(total_death_score)
plt.scatter(X, y_deaths, color = 'magenta', label = 'Original Data')
y_pred = np.array(scalerY4.inverse_transform(y_pred_total_death.reshape(-1,1)))
x_test = np.array(scalerX4.inverse_transform(scaled_X_test_total_deaths))
plt.scatter(x_test, y_pred, color = 'green', label = 'Test Data')
plt.title('Covid19 (Support Vector Regression Model)')
plt.xlabel('Days')
plt.ylabel('Total Deaths')
plt.legend()
plt.ylim([0, 1500])
plt.show()
performance.append(l)

l = []
svr5 = SVR(C=1.0, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma='auto', kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
svr5.fit(scaled_X_train_total_recovery, scaled_y_train_total_recovery)
total_recovery_score = svr5.score(scaled_X_test_total_recovery, scaled_y_test_total_recovery)
y_pred_total_recovery = svr5.predict(scaled_X_test_total_recovery)
total_recovery_MSE = mean_squared_error(scaled_y_test_total_recovery, y_pred_total_recovery)
total_recovery_RMSE = mean_squared_error(scaled_y_test_total_recovery, y_pred_total_recovery, squared = False)
l.append(total_recovery_MSE)
l.append(total_recovery_RMSE)
l.append(total_recovery_score)

plt.scatter(X, y_recovery, color = 'magenta', label = 'Original Data')
y_pred = np.array(scalerY5.inverse_transform(y_pred_total_recovery.reshape(-1,1)))
x_test = np.array(scalerX5.inverse_transform(scaled_X_test_total_recovery))
plt.scatter(x_test, y_pred, color = 'green', label = 'Test Data')
plt.title('Covid19 (Support Vector Regression Model)')
plt.xlabel('Days')
plt.ylabel('Total Recovery')
plt.legend()
plt.ylim([0, 10000])
plt.show()
performance.append(l)

l = []
svr6 = SVR(C=1.0, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma='auto', kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
svr6.fit(scaled_X_train_total_confirmed, scaled_y_train_total_confirmed)
total_confirmed_score = svr6.score(scaled_X_test_total_confirmed, scaled_y_test_total_confirmed)
y_pred_total_confirmed = svr6.predict(scaled_X_test_total_confirmed)
total_confirmed_MSE = mean_squared_error(scaled_y_test_total_confirmed, y_pred_total_confirmed)
total_confirmed_RMSE = mean_squared_error(scaled_y_test_total_confirmed, y_pred_total_confirmed, squared = False)
l.append(total_confirmed_MSE)
l.append(total_confirmed_RMSE)
l.append(total_confirmed_score)
plt.scatter(X, y_confirmed, color = 'magenta', label = 'Original Data')
y_pred = np.array(scalerY6.inverse_transform(y_pred_total_confirmed.reshape(-1,1)))
x_test = np.array(scalerX6.inverse_transform(scaled_X_test_total_confirmed))
plt.scatter(x_test, y_pred, color = 'green', label = 'Test Data')
plt.title('Covid19 (Support Vector Regression Model)')
plt.xlabel('Days')
plt.ylabel('Total Confirmed')
plt.legend()
plt.ylim([0, 35000])
plt.show()
performance.append(l)

performance = pd.DataFrame(performance, columns=('MSE', 'RMSE', 'R2 score'), index=['new death', 'new recovery', 'new confirmed', 'total death', 'total recovery', 'total confirmed'])
print(performance)
```
